# Data Preprocessing

```python
# Applying onehot encoding
df.iloc[:,:-1]=pd.get_dummies(df.iloc[:,:-1],drop_first=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Salary']=le.fit_transform(df['Salary'])
df
```

# Splitting into dependent and independent

```python
x=df.iloc[:,:-1]
x
```

```python
y=df['Salary']
y
```

# Standardizing data

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape,x_test.shape
```

# Model Building

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(svc.score(x_train,y_train),svc.score(x_test,y_test))
```

# Hyperparameter tunning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param={
    'C':[0.1,0.001,1],
    'kernel':['linear','poly','rbf'],
    'gamma':[0.1,0.001,1]
}
```

```python
grid=GridSearchCV(SVC(),param_grid=param,verbose=1)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
svc=SVC(C=0.1,gamma=0.1,kernel='poly')
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
y_pred
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(svc.score(x_train,y_train),svc.score(x_test,y_test))
```
